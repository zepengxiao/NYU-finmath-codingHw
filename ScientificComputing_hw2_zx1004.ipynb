## Problem 1:

Prove the following properties for the matrix norm, where $A, B$ are matrices, $\boldsymbol u$ is a vector and $b$ is a scaler.
$\renewcommand{bs}{\boldsymbol}$
* $\Vert A ^{-1} \Vert = \max_{\bs u \ne 0} \frac{\Vert \bs u\Vert}{\Vert A \bs u \Vert} = \left(\min_{\bs u \ne 0} \frac{\Vert A \bs u \Vert}{\Vert \bs u \Vert}\right)^{-1}$
* $\Vert A \boldsymbol u \Vert  \le \Vert A\Vert \; \Vert \boldsymbol u\Vert $
* $\Vert b A\Vert  = \vert b\vert \; \Vert A\Vert $ 
* $\Vert A + B\Vert  \le \Vert A\Vert  + \Vert B\Vert $
* $\Vert AB\Vert  \le \Vert A\Vert \Vert B\Vert$

### Problem 1 Solution:

We know that for a matrix $A$: $\displaystyle \Vert A \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}$. <br>

(a) <br>

$A^{-1}$ exists $\Rightarrow A$ nonsingular, then we have for all vectors $\boldsymbol u$, $A \boldsymbol u \neq 0 \iff A \neq 0$

$\displaystyle \Vert A^{-1} \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert A^{-1} \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} = \max_{A \boldsymbol u \neq 0} \frac{\Vert A^{-1} (A \boldsymbol u) \Vert}{\Vert A \boldsymbol u \Vert} = \max_{A \boldsymbol u \neq 0} \frac{\Vert \boldsymbol u \Vert}{\Vert A \boldsymbol u \Vert} = \max_{\boldsymbol u \neq 0} \frac{\Vert \boldsymbol u \Vert}{\Vert A \boldsymbol u \Vert}$

$\displaystyle \max_{\boldsymbol u \neq 0} \frac{\Vert \boldsymbol u \Vert}{\Vert A \boldsymbol u \Vert} = \max_{\boldsymbol u \neq 0}\frac{1}{\displaystyle \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}} = \frac{1}{\displaystyle\min_{\boldsymbol u \neq 0}\frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}} = (\min_{\boldsymbol u \neq 0}\frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert})^{-1}$

Therefore, $\displaystyle \Vert A^{-1} \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert \boldsymbol u \Vert}{\Vert A \boldsymbol u \Vert} = (\min_{\boldsymbol u \neq 0}\frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert})^{-1}$

(b) $\boldsymbol u$ is a vector, and $\displaystyle \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} \leq \max_{\boldsymbol v \neq 0} \frac{\Vert A \boldsymbol v \Vert}{\Vert \boldsymbol v \Vert} = \Vert A \Vert$, hence: <br>

$\displaystyle \Vert A \boldsymbol u \Vert = \Vert \boldsymbol u \Vert \; \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} \leq \Vert \boldsymbol u \Vert \; \max_{\boldsymbol v \neq 0} \frac{\Vert A \boldsymbol v \Vert}{\Vert \boldsymbol v \Vert} = \Vert A \Vert \; \Vert \boldsymbol u \Vert$

(c) $b$ is a scalar <br>

$\displaystyle \Vert b A \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert bA \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} = \max_{\boldsymbol u \neq 0} \frac{\vert b \vert \; \Vert A \boldsymbol u \Vert}{\Vert u \Vert} = \vert b \vert \; \max_{\boldsymbol u \neq 0} \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} = \vert b \vert \; \Vert A \Vert$

(d) <br>

$\displaystyle \Vert A+B \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert (A+B) \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} = \max_{\boldsymbol u \neq 0} \frac{\Vert A \boldsymbol u + B \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}$

$\displaystyle \Rightarrow \Vert A+B \Vert \leq \max_{\boldsymbol u \neq 0} \frac{\Vert A \boldsymbol u \Vert + \Vert B \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}$ by triangle inequality for vector norm.

$\displaystyle \Rightarrow \Vert A+B \Vert \leq \max_{\boldsymbol u \neq 0} (\frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} + \frac{\Vert B \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} \leq \max_{\boldsymbol u \neq 0} \frac{\Vert A \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} + \max_{\boldsymbol v \neq 0} \frac{\Vert B \boldsymbol v \Vert}{\Vert \boldsymbol v \Vert} = \Vert A \Vert + \Vert B \Vert$

Therefore, $\displaystyle \Vert A+B \Vert \leq \Vert A \Vert + \Vert B \Vert$

(e) <br>

$\displaystyle \Vert AB \Vert = \max_{\boldsymbol u \neq 0} \frac{\Vert AB \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} \leq \max_{\boldsymbol u \neq 0} \frac{\Vert A \Vert \; \Vert B \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}$ (Consider $\boldsymbol v = B \; \boldsymbol u$, then $\Vert A \boldsymbol v \Vert  \le \Vert A\Vert \; \Vert \boldsymbol v\Vert $ by property (b)).

$\displaystyle \Rightarrow \Vert AB \Vert \leq \Vert A \Vert \; \max_{\boldsymbol u \neq 0} \frac{\Vert B \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert} \leq \Vert A \Vert \; \max_{\boldsymbol v \neq 0} \frac{\Vert B \boldsymbol v \Vert}{\Vert \boldsymbol v \Vert} = \Vert A\Vert \Vert B\Vert$, where $\boldsymbol u$ leads to $\displaystyle \max_{\boldsymbol u \neq 0} \frac{\Vert AB \boldsymbol u \Vert}{\Vert \boldsymbol u \Vert}$ and $\boldsymbol v$ leads to $\displaystyle \max_{\boldsymbol v \neq 0} \frac{\Vert B \boldsymbol v \Vert}{\Vert \boldsymbol v \Vert}$.

Therefore, $\displaystyle \Vert AB \Vert \leq \Vert A\Vert \Vert B\Vert$

### Problem 2 Solution:

1. <br>

Suppose that $\left(\begin{array} \\ 1 & \rho \\ \rho & 1 \end{array} \right) = LL^{T}$, where $L = \left(\begin{array} \\ a_1 & 0 \\ a_2 & a_3 \end{array} \right)$ is the lower triangular martrix.

Then $LL^{T} = \left(\begin{array} \\ a_1 & 0 \\ a_2 & a_3 \end{array} \right)\left(\begin{array} \\ a_1 & a_2 \\ 0 & a_3 \end{array} \right) = \left(\begin{array} \\ a_1^2 & a_1a_2 \\ a_1a_2 & a_2^2 + a_3^2 \end{array} \right)$

Hence we have: $a_1^2 = 1, a_1a_2 = \rho, a_2^2 + a_3^2 = 1 \Rightarrow a_1 = 1, a_2 = \rho, a_3 = \sqrt{1 - \rho^2} \Rightarrow L = \left(\begin{array} \\ 1 & 0 \\ \rho & \sqrt{1 - \rho^2} \end{array} \right)$

Therefore, the analytic expression of Cholesky decomposition of $\left(\begin{array} \\ 1 & \rho \\ \rho & 1 \end{array} \right)$ is: <br>

$$\left(\begin{array} \\ 1 & \rho \\ \rho & 1 \end{array} \right) = \left(\begin{array} \\ 1 & 0 \\ \rho & \sqrt{1 - \rho^2} \end{array} \right)\left(\begin{array} \\ 1 & \rho \\ 0 & \sqrt{1 - \rho^2} \end{array} \right)$$ <br>

2. <br> 

The recursive Cholesky decomposition algorithm discussed in the class is implemented below:

Suppose that the input matrix is $M \in \mathbb{R^{n\times n}}$, we use $m_{11} <= 0$ to check if the program may fail(if the matrix is SPD or not), since we need $l_{11} = \sqrt{m_{11}}$, and for the recursive part, we need $L_{21} = \frac{M_{21}}{l_{11}}$.

````python
#Implement the recursive Cholesky decomposition algorithm
#Input:
#  M: The matrix that need to do Cholesky decomposition
#Output:
#  L: The lower triangular matirx such that M = LL^T
def Chol(M):
    # Check if the input matrix is semi positive definite (SPD)
    # The whole part of this program may fail if m_{11} < 0
    if (M[0, 0] < 0):
        raise RuntimeError('Input matrix is not SPD.')

    dim = len(M)
    # base case: if M is 1x1 matirx, then directly return sqrt(M)
    if (dim == 1):
        L = np.sqrt(M)
        return L
    # recursive part, L contanis three parts: l_{11}, L_{21} and L_{22}
    else:
        # Check if the input matrix is semi positive definite (SPD)
        # The recursive part of this program may fail if m_{11} <= 0
        if (M[0, 0] == 0):
            raise RuntimeError('Input matrix is not SPD.')
  
        L = np.zeros((dim, dim)) # Initialize the result matrix L
        L[0, 0] = np.sqrt(M[0, 0]) # l_{11} = \sqrt{m_{11}}
        L[1:,0:1] = M[1:,0:1] / L[0, 0] # L_{21} = M_{21}/l_{11}
        # L_{22} is the Cholesky decomposition of (M_{22} - (M_{21}M_{21}^T)/m_{11})
        L[1:,1:] = Chol(M[1:,1:] - np.matmul(M[1:,0:1], M[1:,0:1].T)/M[0, 0])
        return L
````

````python
import time

# Initialize matrix dimension and the array to record the time 
# of doing Cholesky decomposition for different size of matrix
dim = np.arange(1, 101)
handTime = np.zeros(100)
scipyTime = np.zeros(100)

# Recording time for different size of matrix
for i in range(100):
    d = dim[i]
    # Make sure that the matrix is SPD
    M = np.matmul(np.tril(np.arange(1, d + 1)), np.tril(np.arange(1, d + 1)).T)
    
    # Recording time for self-implemented Cholesky
    handStart = time.clock()
    handL = Chol(M)
    handEnd = time.clock()
    handTime[i] = handEnd - handStart
    
    # Recording time for scipy.linalg
    scipyStart = time.clock()
    scipyL = scipy.linalg.cholesky(M, lower = True)
    scipyEnd = time.clock()
    scipyTime[i] = scipyEnd - scipyStart
    
    assert(np.allclose(handL, scipyL))

# Plot results
plt.plot(dim, handTime, label = 'Self-implemented recursive method')
plt.plot(dim, scipyTime, label = 'Scipy.linalg.Cholesky')
plt.plot(dim, handTime - scipyTime, label = 'Time Difference')
plt.title('Performance between recursive method and scipy.linalg')
plt.xlabel('Matrix Size')
plt.ylabel('Computaional Time')
plt.legend()
````

3. <br>

The comparison of performance between self-implemented program and scipy.linalg.Cholesky is showed above.

We may conclude that scipy.linalg.Cholesky has better performance, and the difference of performance is increasing as the size of matrix increasing.

### Problem 3 Solution:

````python
# Work for question 1

histReturn = np.log(histprice).diff() # daily returns
histCorr = histReturn.corr() # correlation matrix
histCov = histReturn.cov() # covariance matrix
# annualize and display the result, assume 252 business days per year
fmt.displayDFs(histCorr, headers=['Correlation'], fontsize=4, fmt="4f")
fmt.displayDFs(histCov*252, headers=['Covariance'], fontsize=4, fmt="4f")
````

1. <br>

The historical correlation and covariance matrix of the DJIA constituents' daily returns are showed above.

````python
# Work for question 2

rf = np.float64(0.03) # annualized historical risk-free rate
histExcessReturn = np.mean(histReturn)*252*(-1) - rf # annualize historical excess return
# normalize historical excess return since it is only meaningful in relative sense
histExcessReturnNormal = histExcessReturn / np.sum(histExcessReturn) 
df_Result = pd.DataFrame(histExcessReturn*100, columns=['Excess Return %']).transpose() # convert the result to a dataframe
# display the result
df_Result.loc['Normalized Excess Return %'] = histExcessReturnNormal*100
fmt.displayDF(df_Result[:2], fontsize=4, fmt="4g")
````

2. <br>

The historical excess return of individual stocks is showed above.

````python
# Work for question 3

inv_histCov = np.linalg.inv(histCov) # inverse of covariance matrix C^{-1}
array_histExcessReturn = np.array(histExcessReturn) # convert panda series to np.array to do matrix operations
# optimal portfolio weight = (C^{-1}*f)/(f^T*C^{-1}*f)
wHist = inv_histCov.dot(array_histExcessReturn.T)/array_histExcessReturn.T.dot(inv_histCov).dot(array_histExcessReturn)
wHist = wHist/np.sum(wHist) # normalized optimal portfolio weight
# display the result
df_Result.loc['Optimal Portfolio(O)', :] = wHist
fmt.displayDF(df_Result[-1:], "4g", fontsize=4)
````

3. <br>

The optimal portfolio using the estimated historical excess return and covariance matrix is showed above.

From the excess return we have in question2, all stocks should have positive return or negative return simultaneously, which means that the normalized weight of each stock should be greater than or equal to zero. If we believe that one stock has positive excess return and buy it, then so are the other stocks, we may not buy the others but we have no reason to sell them

````python
# Work for Qeustion 6

# Historical correlation and covariance matrix
dataurl = "https://raw.githubusercontent.com/yadongli/nyumath2048/master/data/djiahist.csv"
histprice2 = pd.read_csv(dataurl, index_col=[0])
histprice2.sort_index()
histprice2['date'] = histprice2.index
histprice2['date'] = pd.to_datetime(histprice2['date'])
histprice2['week'] = histprice2['date'].dt.week
histprice2['year'] = histprice2['date'].dt.year
histprice2 = histprice2.drop(['date'], axis=1)
# convert daily price to weekly price
weekPrice = histprice2.groupby(['year','week']).mean()
weekReturn = np.log(weekPrice).diff()
weekCorr = weekReturn.corr() # correlation matrix
weekCov = weekReturn.cov() # covariance matrix
# annualize and display the result, assume 52 business weeks per year
fmt.displayDFs(weekCorr, headers=['Correlation'], fontsize=4, fmt="4f")
fmt.displayDFs(weekCov*52, headers=['Covariance'], fontsize=4, fmt="4f")
````

````python
# Historical excess return

rf = np.float64(0.03) # annualized historical risk-free rate
weekExcessReturn = np.mean(weekReturn)*(-1)*52 - rf # annualize historical excess return
# normalize historical excess return since it is only meaningful in relative sense
weekExcessReturnNormal = weekExcessReturn / np.sum(weekExcessReturn) 
df_weekResult = pd.DataFrame(weekExcessReturn*100, columns=['Excess Return %']).transpose() # convert the result to a dataframe
# display the result
df_weekResult.loc['Normalized Excess Return %'] = weekExcessReturnNormal*100
fmt.displayDF(df_weekResult[:2], fontsize=4, fmt="4g")
````

````python
# Optimal portfolio using the estimated historical excess return and covariance matrix

inv_weekCov = np.linalg.inv(weekCov) # inverse of covariance matrix C^{-1}
array_weekExcessReturn = np.array(weekExcessReturn) # convert panda series to np.array to do matrix operations
# optimal portfolio weight = (C^{-1}*f)/(f^T*C^{-1}*f)
wWeek = inv_weekCov.dot(array_weekExcessReturn.T)/array_weekExcessReturn.T.dot(inv_weekCov).dot(array_weekExcessReturn)
wWeek = wWeek/np.sum(wWeek) # normalized optimal portfolio weight
# display the result
df_weekResult.loc['Optimal Portfolio(O)', :] = wWeek
fmt.displayDF(df_weekResult[-1:], "4g", fontsize=4)
````

````python
# Implied return forecast

wWeekActual = np.array(djia['Weights']) # actual weights from actual DJIA portfolio w_b
# beta = (C*w_b)/(sigma_b^2)
impliedWeekReturnActual = weekCov.dot(wWeekActual)/(wWeekActual.dot(weekCov).dot(wWeekActual))
# normalize implied return since it is only meaningful in relative sense
impliedWeekReturnActualNormal = impliedReturnActual / np.sum(impliedReturnActual)
# display the result
df_weekResult.loc['Implied Return %', :] = impliedWeekReturnActual*100
df_weekResult.loc['Normalized Implied Return %'] = impliedWeekReturnActualNormal*100
fmt.displayDF(df_weekResult[-2:], "4g", fontsize=4)
````

````python
# Condition number

# Use k(A) = ||A||*||A^{-1}|| to compute the condition number for matrix A
condWeekCov = np.matrix([np.linalg.norm(weekCov, n)*np.linalg.norm(inv_weekCov, n) for n in (1, 2, inf)])
# display the result
fmt.displayDF(pd.DataFrame(condWeekCov, columns = ["L-1", "L-2", "L-$\infty$"], index=['Condition number']), 
              "4g", fontsize=4)
````

````python
# Work for Qeustion 8

np.random.seed(11078211)

# compute the actual Sharpe ratio by historical excess return and covariance
sharpeRatio = wHist.T.dot(np.array(histExcessReturnNormal)) / np.sqrt((wHist.T.dot(histCov*252).dot(wHist)))
# initialize an array to store the Sharpe ratio of optimal portfolio with pertubation
srPertubations = [0] * 10

# do 10 pertubations
for i in range(10):
    wHistPer = wHist
    # for each pertubation, randomly choose the number of(1-5) weights to give pertubation
    times = np.random.randint(1, 6)
    for j in range(times):
        # randomly choose weight
        index = np.random.randint(0, len(wHist))
        # for each weight that is pertubated, it has either positive pertubation or negative pertubation
        wHistPer[index] *= (1 + 1e-3 * (np.random.randint(1, 6) * 2 - 1))
    # store the Sharpe ratio of this pertubation
    srPertubations[i] = wHistPer.T.dot(np.array(histExcessReturnNormal)) / np.sqrt((wHistPer.T.dot(histCov*252).dot(wHistPer)))

print("The actual Sharpe ratio is:", sharpeRatio)
print("The Sharpe ratio of optimal portfolio with different pertubations are:", srPertubations)
print("The actual Sharpe ration is greater than all pertubated Sharpe ratio is", (srPertubations < sharpeRatio).all())
````

9. <br>

Sharpe ratio is a great way to measure risk-adjusted-performance(considering both return and risk), since the optimal portfolio is computed by mean-variance model whose parameters are historical excess return and covariance, and it exactly has greatest sharpe ratio.

However, the Sharpe ratio may not be reliable if the forecast of future return or volatility is not accurate. For example here, the optimal portfolio is not reasonable though it has the greatest sharpe ratio, since estimating future return by historical excess return is incorrect.

Therefore, when tend to Sharpe ratio to show the performance, parameters in Sharpe ratio like return vector and covariance matrix, should be justified first.
